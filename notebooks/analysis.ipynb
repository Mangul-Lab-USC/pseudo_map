```python
import pandas as pd
import seaborn as sns
```

```python
df_50_100=pd.read_csv('../raw_data/df_length_50_100.csv')
df_50_100.head()
```

```python
df_100_500=pd.read_csv('../raw_data/df_length_100_500.csv')
df_100_500.head()
```

```python
df_500_1000=pd.read_csv('../raw_data/df_length_500_1000.csv')
df_500_1000.head()
```

```python
df_1000=pd.read_csv('../raw_data/df_length_1000_.csv')
df_1000.head()
```

```python
vertical_stack = pd.concat([df_50_100, df_100_500,df_500_1000,df_1000], axis=0)
vertical_stack.tail()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(40, 16))


g=sns.set_style("white")
g=sns.set_context("talk")
g = sns.barplot(x="tool", y="precision",data=vertical_stack,hue='len', ci=None)
# g.set_xticklabels(rotation=90)
g.set(xlabel='', ylabel='precision')

for p in g.patches:
    g.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2.,
                                           p.get_height()),ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
g=sns.despine()


```

```python
df_best_50_100 = vertical_stack.loc[(vertical_stack["tool"] == "smoove") & (vertical_stack["len"] == "50_100")]

df_best_100_500 = vertical_stack.loc[(vertical_stack["tool"] == "smoove") & (vertical_stack["len"] == "100_500")]

df_best_500_1000 = vertical_stack.loc[(vertical_stack["tool"] == "breakdancer") & (vertical_stack["len"] == "500_1000")]

df_best_1000 = vertical_stack.loc[(vertical_stack["tool"] == "clever") & (vertical_stack["len"] == "1000_")]


combined_best_tool = pd.concat([df_best_50_100, df_best_100_500,df_best_500_1000,df_best_1000], axis=0)

combined_best_tool["tool"] = "pseudo_tool"

combined = pd.concat([combined_best_tool,vertical_stack ], axis=0)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(40, 16))


g=sns.set_style("white")
g=sns.set_context("talk")
g = sns.barplot(x="tool", y="f-score",data=combined,hue='len', ci=None)
# g.set_xticklabels(rotation=90)
g.set(xlabel='', ylabel='f-score')

for p in g.patches:
    g.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2.,
                                           p.get_height()),ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
g=sns.despine()
```
